# load necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Basic Graph 
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#passing label
plt.plot(x,y,label='2x',color='red')

# To change font name and font size in graph title
plt.title("Graph Practice",fontdict = {'fontname':'Comic Sans MS',
                                'fontsize':20,'color': 'purple'})

#change X-axis font size ,color,name 
plt.xlabel("X axis",
           fontdict={'fontname':'Arial Black','fontsize':15,'color':'b'})

#change Y-axis font size ,color,name 
plt.ylabel("Y axis",
           fontdict={'fontname':'Arial Black','fontsize':15,'color':'r'})
#Ticks in graph 
plt.xticks([0, 1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()

plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#passing label
#and use of marker ,label , linewidth

plt.plot(x,y,label='2x',color ='purple',
         linewidth=2,
         marker = '.',
         markersize =10,
         markeredgecolor='black')

plt.legend() 
```

# When your graph is stop and you want to show the projection of it not in the future what i did to show that type of phenomenon 

* X to take first 6 values and plot that
* X to take last 4 values and plot that


```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
#Resize your graph and also use below the X & Y values array 
plt.figure(figsize=(5,3),dpi=300)

#line number two

x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,'r',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'r--')


#'r ^ --'= '[color(r)] [marker(^)] [line(--)]'
#this line is for first line in graph red color 
plt.plot(x,y,'b^--',label ='2x')

#X,Y ticksmark

plt.xticks([0, 1,2,3,4])
#plt.yticks([0,2,4,6,8,10])
plt.savefig('mygraph.png',dpi=300)
plt.legend()
plt.show()
```

# Bar Chart

```python
labels =['a','b','c']
values =[1,4,2]

bar = plt.bar(labels,values)

#for lines in the bar column[0]

bar[0].set_hatch('/') 
bar[2].set_hatch('*')
plt.figure(figsize=(6,4))


plt.show() 
```

# Alternative of following code:

```python
labels =['a','b','c']
values =[1,4,2]

bar1 = plt.bar(labels,values)

#for lines in the bar column
patterns = ['/','o','*']
for bar in bar1:
    bar.set_hatch(patterns.pop(0))
    
plt.figure(figsize=(6,4))
plt.show() 
```
