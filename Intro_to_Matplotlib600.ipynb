#Line Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
growth_rate = np.array([5,5.2,6,7.3,9,8.5,12])
```

```python
plt.plot(growth_rate)
```

```python
plt.plot(growth_rate,'o-')
```

```python
plt.plot(growth_rate,'o--')
```

```python
plt.plot(growth_rate,'o-',color = 'purple')
```

```python
plt.plot(growth_rate,'o-',color = 'purple',ms = 8,linewidth = 5) #here 'ms = 8 ' is marker size & 'linewidth = 5' is thickness of line
```

```python
#Add title on plot
plt.plot(growth_rate,'o-',color = 'purple')
plt.title("Growth rate for each month")
```

```python
#to remove this "Text(0.5, 1.0, 'Growth rate for each month" use 'plt.show()'

plt.plot(growth_rate,'o-',color = 'purple')
plt.title("Growth rate for each month")
plt.show()
```

```python
growth_rate = np.array([5,5.2,6,7.3,9,8.5,12])
predicted_growth = np.array([5.2,5.5,5.9,8,10,9,11.7])

plt.plot(growth_rate,'o-',color = 'r',label = 'Predicted Growth')

plt.plot(growth_rate,'o-',color = 'purple',label = 'Actual Growth')

plt.title("Growth rate for each month")
plt.xlabel("Months")
plt.ylabel("Growth rate(%)")
plt.legend()
plt.show()
```
